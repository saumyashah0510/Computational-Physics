## Mathematical Analysis: Energy Conservation in Euler-Cromer

### Part 1: Derivation of the Energy Difference Equation

For a simple harmonic oscillator, the differential equation is $\ddot{x} = -x$. This gives us the acceleration $a = -x$.

The **Euler-Cromer** update rules (which update velocity *first*, and then use the new velocity to update position) are defined as:
1. $v_{n+1} = v_n - x_n \Delta t$
2. $x_{n+1} = x_n + v_{n+1} \Delta t$

We define the true physical energy of the system as $E_n = \frac{1}{2}(v_n^2 + x_n^2)$. 
Our goal is to find the difference $E_{n+1} - E_n = \frac{1}{2}(v_{n+1}^2 + x_{n+1}^2) - \frac{1}{2}(v_n^2 + x_n^2)$. 

**Step 1: Expand $v_{n+1}^2$**
Using rule 1, we square the new velocity:
$$v_{n+1}^2 = (v_n - x_n \Delta t)^2$$
$$v_{n+1}^2 = v_n^2 - 2v_n x_n \Delta t + x_n^2 (\Delta t)^2$$

**Step 2: Expand $x_{n+1}^2$**
First, substitute the expression for $v_{n+1}$ into rule 2 so everything is in terms of step $n$:
$$x_{n+1} = x_n + (v_n - x_n \Delta t) \Delta t$$
$$x_{n+1} = x_n + v_n \Delta t - x_n (\Delta t)^2$$

Now, we square this trinomial using the expansion formula $(A + B - C)^2 = A^2 + B^2 + C^2 + 2AB - 2AC - 2BC$:
* Let $A = x_n$
* Let $B = v_n \Delta t$
* Let $C = x_n (\Delta t)^2$

$$x_{n+1}^2 = (x_n)^2 + (v_n \Delta t)^2 + (x_n (\Delta t)^2)^2 + 2(x_n)(v_n \Delta t) - 2(x_n)(x_n (\Delta t)^2) - 2(v_n \Delta t)(x_n (\Delta t)^2)$$

Simplify the terms:
$$x_{n+1}^2 = x_n^2 + v_n^2 (\Delta t)^2 + x_n^2 (\Delta t)^4 + 2v_n x_n \Delta t - 2x_n^2 (\Delta t)^2 - 2v_n x_n (\Delta t)^3$$

**Step 3: Combine to find the sum of squares $(v_{n+1}^2 + x_{n+1}^2)$**
Add the expanded forms of $v_{n+1}^2$ and $x_{n+1}^2$ together:
$$(v_{n+1}^2 + x_{n+1}^2) = \left[ v_n^2 - \underline{2v_n x_n \Delta t} + x_n^2 (\Delta t)^2 \right] + \left[ x_n^2 + v_n^2 (\Delta t)^2 + x_n^2 (\Delta t)^4 + \underline{2v_n x_n \Delta t} - 2x_n^2 (\Delta t)^2 - 2v_n x_n (\Delta t)^3 \right]$$

Notice that the $-2v_n x_n \Delta t$ and $+2v_n x_n \Delta t$ terms completely cancel each other out. 
Gathering the remaining terms by their powers of $\Delta t$:
$$(v_{n+1}^2 + x_{n+1}^2) = v_n^2 + x_n^2 + [x_n^2 + v_n^2 - 2x_n^2](\Delta t)^2 - 2v_n x_n (\Delta t)^3 + x_n^2(\Delta t)^4$$

Simplify the $(\Delta t)^2$ bracket:
$$(v_{n+1}^2 + x_{n+1}^2) = (v_n^2 + x_n^2) + (v_n^2 - x_n^2)(\Delta t)^2 - 2v_n x_n (\Delta t)^3 + x_n^2(\Delta t)^4$$

**Step 4: Multiply by $\frac{1}{2}$ to find $E_{n+1} - E_n$**
Since $E_n = \frac{1}{2}(v_n^2 + x_n^2)$, we multiply our combined equation by $\frac{1}{2}$ and subtract $E_n$:
$$E_{n+1} - E_n = \frac{1}{2} \left[ (v_n^2 - x_n^2)(\Delta t)^2 - 2v_n x_n(\Delta t)^3 + x_n^2(\Delta t)^4 \right]$$

### Part 2: Averaging Over One Cycle

To understand why Euler-Cromer is so stable for oscillatory systems, we must look at how these error terms behave over a full period (cycle) of simple harmonic motion.

The analytical, exact solutions for simple harmonic motion $\ddot{x} = -x$ are:
* Position: $x(t) = \cos(t)$
* Velocity: $v(t) = -\sin(t)$

Let's evaluate the coefficients of $(\Delta t)^2$ and $(\Delta t)^3$ by substituting these exact solutions, and integrate them over one full period $T = 2\pi$ to find their average value.

**1. The Coefficient of $(\Delta t)^2$:**
The base coefficient is $(v_n^2 - x_n^2)$.
Substituting our analytical functions:
$$v(t)^2 - x(t)^2 = (-\sin t)^2 - (\cos t)^2 = \sin^2 t - \cos^2 t$$
Using the double-angle trigonometric identity $\cos(2t) = \cos^2 t - \sin^2 t$, we can rewrite this as:
$$v(t)^2 - x(t)^2 = -\cos(2t)$$

Now, take the average of $-\cos(2t)$ over one cycle ($0$ to $2\pi$):
$$\text{Average} = \frac{1}{2\pi} \int_{0}^{2\pi} -\cos(2t) \, dt$$
$$\text{Average} = \frac{-1}{2\pi} \left[ \frac{\sin(2t)}{2} \right]_{0}^{2\pi} = \frac{-1}{4\pi} (\sin(4\pi) - \sin(0)) = \frac{-1}{4\pi} (0 - 0) = \mathbf{0}$$

**2. The Coefficient of $(\Delta t)^3$:**
The base coefficient is $-2v_n x_n$.
Substituting our analytical functions:
$$-2v(t)x(t) = -2(-\sin t)(\cos t) = 2\sin t \cos t$$
Using the double-angle identity $\sin(2t) = 2\sin t \cos t$, we rewrite this as:
$$-2v(t)x(t) = \sin(2t)$$

Take the average over one cycle:
$$\text{Average} = \frac{1}{2\pi} \int_{0}^{2\pi} \sin(2t) \, dt$$
$$\text{Average} = \frac{1}{2\pi} \left[ \frac{-\cos(2t)}{2} \right]_{0}^{2\pi} = \frac{-1}{4\pi} (\cos(4\pi) - \cos(0)) = \frac{-1}{4\pi} (1 - 1) = \mathbf{0}$$

### Conclusion

Because the mathematical averages of the $\mathcal{O}(\Delta t^2)$ and $\mathcal{O}(\Delta t^3)$ error terms are exactly **zero** over a full oscillation, the energy errors that accumulate during one half of the cycle are perfectly canceled out by errors with the opposite sign during the second half of the cycle. 

Therefore, the net change in energy over one complete period is dominated solely by the remaining term, $\frac{1}{2}x_n^2(\Delta t)^4$. Because $(\Delta t)$ is small, $(\Delta t)^4$ is incredibly tiny, proving that the **Euler-Cromer method conserves energy up to $\mathcal{O}(\Delta t^4)$** for simple harmonic motion.
